```python
import matplotlib.pyplot as plt
```

```python
def open_file(file_name):
    with open(file_name, 'r') as data:
        xx = []
        yy = []
        for line in data:
            p = line.split()
            xx.append(float(p[0]))
            yy.append(float(p[1]))

    return xx, yy

xx, yy = open_file('out.txt')
plt.plot(xx,yy)
```
