```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def open_file(file_name):
    with open(file_name, 'r') as data:
        xx = []
        yy = []
        for line in data:
            p = line.split()
            xx.append(float(p[0]))
            yy.append(float(p[1]))

    return np.array(xx), np.array(yy)

plt.figure(figsize=(16,9))
plt.xscale("log")
plt.xlabel('Picture size')
plt.ylabel('Time of execution (s)')
plt.title('Comparison of different approaches to parallel programming')
for filename in os.listdir("outputs"):
    xx, yy = open_file(os.path.join("outputs", filename))
    style = '-'
    if "pools" in filename:
        style='--'
    elif"threads" in filename:
        style='-.'
    plt.plot(xx**2,yy,style,label=filename)

plt.grid()
plt.legend()
```
